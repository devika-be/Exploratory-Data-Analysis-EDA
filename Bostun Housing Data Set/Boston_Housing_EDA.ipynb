#### Finding and Removing Outliers:

```python
fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
index = 0
axs = axs.flatten()
for k,v in data.items():
    sns.boxplot(y=k, data=data, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

```python
for k, v in data.items():
    # k - Feature , v - values from that feature
    
    q1 = v.quantile(0.25) 
    q3 = v.quantile(0.75)
    irq = q3 - q1 # Inter Quartile Range

    #  | - OR 
    # we are creating list of elements for particular features
    
    v_col = v[(v <= q1 - 1.5 * irq) | (v >= q3 + 1.5 * irq)]
    
    
    perc = np.shape(v_col)[0] * 100.0 / np.shape(data)[0]
    
    print("Column %s outliers = %.2f%%" % (k, perc))
```

Q2. Do We have to pay more to get house near Charles river?

```python
import pandas as pd
from scipy import stats

# null hypothesis: There is no relation between price and house near river
# Alternate hypothesis: There is relation between price and house near river

# Create two groups: houses near river and houses not near river
river_prices = data.loc[data['CHAS'] == 1]['MEDV'] # data.loc[data['CHAS'] == 1] we are using this part as labeled index.
not_river_prices = data.loc[data['CHAS'] == 0]['MEDV']

# Perform a two-sample t-test assuming unequal variances
t_statistic, p_value = stats.ttest_ind(river_prices, not_river_prices, equal_var=False)
print('t_statistic: ', t_statistic)
print('p_value: ', p_value)
# Check if p-value is less than 0.05 (assuming a significance level of 0.05)
if p_value < 0.05:
    print('There is a significant difference between house prices near and not near the river.') # reject null
else:
    print('There is no significant difference between house prices near and not near the river.') # fail to reject null
```

Q4. Are there any significant differences in the distribution of features between neighborhoods in Boston?

```python
import pandas as pd
from scipy import stats


# Create a list of the feature names to test
features_to_test = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']

# Create a dictionary to store the F-statistics and p-values for each feature
anova_results = {}

# Loop through each feature and perform ANOVA test
for feature in features_to_test:
    # Create a list of the feature values for each neighborhood
    neighborhoods = data['RAD'].unique()
    groups = [data.loc[data['RAD'] == n][feature] for n in neighborhoods]
    
    # Perform ANOVA test
    f_statistic, p_value = stats.f_oneway(*groups)
    
    # Store results in dictionary
    anova_results[feature] = {'F-statistic': f_statistic, 'p-value': p_value}

# Print ANOVA results for each feature
for feature, results in anova_results.items():
    print(f"{feature}: F-statistic = {results['F-statistic']}, p-value = {results['p-value']}")
```

Q4. Is there a relationship between the proportion of non-retail business acres per town (INDUS) and housing prices (MEDV)?

```python
import pandas as pd
import seaborn as sns

# Plot a scatterplot with regression line
sns.regplot(x='INDUS', y='MEDV', data= data)

# Calculate the Pearson correlation coefficient
correlation_coefficient = data['INDUS'].corr(data['MEDV'])

# Print the Pearson correlation coefficient
print("Pearson correlation coefficient between INDUS and MEDV:", correlation_coefficient)
```

Based on the Pearson correlation coefficient of -0.60 between INDUS and MEDV, we can conclude that there is a moderately strong negative correlation between the proportion of non-retail business acres per town and housing prices in the Boston dataset. This means that as the proportion of non-retail business acres per town increases, the housing prices tend to decrease.
